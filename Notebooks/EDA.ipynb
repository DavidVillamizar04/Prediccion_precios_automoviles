```python
# Se importan las librerias a usar
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Se guarda en un dataframe los datos a usar
autos = pd.read_csv('ML_cars.csv')
autos.head(5)
```

```python
# Vemos la informacion del dataset
autos.info()
```

```python
# Guardamos en una variable unicamente los datos numericos del dataframe
datos_numericos= autos.select_dtypes(include=[np.number])
# Graficamos un histograma para ver graficamente las cantidades de vehiculos segun precios
plt.figure(figsize=(10,4))
sns.set_style("white")
sns.histplot(data= datos_numericos, x= 'price', kde= True, color='g')
plt.title('Histograma de precios')
plt.ylabel('Numero de Carros')
plt.xlabel('Precios')
plt.tight_layout()
```

```python
# Graficamos la matriz de correlacion 
plt.figure(figsize=(10, 6))
sns.heatmap(datos_numericos.corr(), annot=True, cmap='flare', linewidths=.5)
plt.title('Matriz de correlaciÃ³n', fontweight='bold', fontsize=20, ha='center', va='bottom')
plt.show()
```

```python
# guardamos los datos de las tres columnas que tuvieron mas correlasion con respecto al precio
columnas = datos_numericos[['horsepower','enginesize','curbweight' ]]

# Realizamos un grafico de dispersion para cada variable que guardamos anteriormente
plt.figure(figsize=(16, 6))
sns.set_style("white")
# Se realiza un ciclo para iterar la lista de variables y colocarlas en cada posicion del grafico ya definido
for i, column in enumerate(columnas):
    plt.subplot(1, 3, i+1)
    sns.scatterplot(data=datos_numericos, x=column, y='price', alpha= 0.8, color= '#80BFFF')
    plt.title(f'{column} vs Precio', fontsize=10)
    plt.grid(True)
    plt.ylabel('Precio')
    plt.tight_layout()
    plt.suptitle('correlacion con el precio', fontsize=20, fontweight="bold")

plt.show()
```
